```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the dataset
df = pd.read_csv("/content/worldpopulationdata.csv")
```

```python
# Display dataset information
print("Dataset Info:")
print(df.info())
```

```python
# Display the first few rows of the dataset
print("\nFirst 5 Rows:")
print(df.head())
```

```python
# Select the top 10 most populated countries in 2022
top_countries = df.nlargest(10, '2022')[['Country Name', '2022']]
```

```python
# Create a bar chart for the top 10 most populated countries in 2022
plt.figure(figsize=(12, 6))
sns.barplot(x='2022', y='Country Name', data=top_countries, hue='Country Name', palette='viridis', legend=False)
plt.xlabel('Population in 2022')
plt.ylabel('Country Name')
plt.title('Top 10 Most Populated Countries in 2022')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

```python
# Create a histogram to show the distribution of population in 2022
plt.figure(figsize=(10, 5))
sns.histplot(df['2022'], bins=30, kde=True, color='blue')
plt.xlabel('Population in 2022')
plt.ylabel('Number of Countries')
plt.title('Distribution of Country Populations in 2022')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
